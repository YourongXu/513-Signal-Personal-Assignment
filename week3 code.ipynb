```python
import numpy as np
import matplotlib.pyplot as plt

N = 6
x = np.array([1, 1, 1, 1, 0, 0])

X = np.fft.fft(x) / N

plt.figure(figsize=(12, 6))

plt.subplot(2, 1, 1)
plt.stem(np.abs(X))
plt.title('Magnitude of DTFS Coefficients')
plt.xlabel('k')
plt.ylabel('|X[k]|')

plt.subplot(2, 1, 2)
plt.stem(np.angle(X))
plt.title('Phase of DTFS Coefficients')
plt.xlabel('k')
plt.ylabel('âˆ X[k]')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

N = 6
x = np.array([1, 1, 1, 1, 0, 0])

X = np.fft.fft(x) / N

n = np.arange(N) 
x_reconstructed = np.zeros(N, dtype=complex) 

for k in range(N):
    x_reconstructed += X[k] * np.exp(1j * 2 * np.pi * k * n / N)

x_reconstructed = np.round(x_reconstructed, 10)

plt.figure(figsize=(12, 6))

plt.subplot(2, 1, 1)
plt.stem(n, x)
plt.title('Original Signal')
plt.xlabel('n')
plt.ylabel('x[n]')

plt.subplot(2, 1, 2)
plt.stem(n, x)
plt.title('Reconstructed Signal')
plt.xlabel('n')
plt.ylabel('x_reconstructed[n]')

plt.tight_layout()
plt.show()
```
